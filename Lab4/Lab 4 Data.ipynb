```python
#First Observing Night
print("First Observing Night")

first_flatclear_list = glob.glob("lab4_flatclear*FIT")
first_flatclear_data = []
for fn in first_flatclear_list:
    data = fits.getdata(fn)
    first_flatclear_data.append(data)
print("Dome Flat clear mean:", np.mean(first_flatclear_data))

first_flatB_list = glob.glob("lab4_flatB*FIT")
first_flatB_data = []
for fn in first_flatB_list:
    data = fits.getdata(fn)
    first_flatB_data.append(data)
print("Dome Flat B mean:", np.mean(first_flatB_data))

first_flatV_list = glob.glob("lab4_flatV*FIT")
first_flatV_data = []
for fn in first_flatV_list:
    data = fits.getdata(fn)
    first_flatV_data.append(data)
print("Dome Flat V mean:", np.mean(first_flatV_data))

first_flatI_list = glob.glob("lab4_flatI*FIT")
first_flatI_data = []
for fn in first_flatI_list:
    data = fits.getdata(fn)
    first_flatI_data.append(data)
print("Dome Flat I mean:", np.mean(first_flatI_data))

first_dark_list = glob.glob("lab4_dark*FIT")
first_dark_data = []
for fn in first_dark_list:
    data = fits.getdata(fn)
    first_dark_data.append(data)
print("Dark (30 seconds) mean:", np.mean(first_dark_data))
```
